```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense, Flatten
```

```python
(X_train, y_train),(X_test, y_test) = keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train[0]
```

```python
# actal (28x28) we have to flatten it so that all the row data will
# be in a single row  (1x784)
# this we have done because we have to provide 784
# inputs parameters to the dense layer
# 28 x 28 = 784
model = Sequential() #Keras Sequential Model
model.add(Flatten(input_shape=(28,28)))
model.add(Dense(128, activation='relu'))
model.add(Dense(10, activation='softmax'))
```

```python
model.summary()
```

```python
# if loss was categorical_crossentropy then we would the convert inputs to one hot encoding
model.compile(loss='sparse_categorical_crossentropy', optimizer='Adam')
```

```python
model.fit(X_train, y_train, epochs=10, validation_split=0.2)
```

```python
y_prob = model.predict(X_test) #will check the probability on the image
```

```python
y_prob
```

```python
y_pred = y_prob.argmax(axis=1)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```
